```python
import os
import shutil
import pandas as pd

# 路径配置
folder_A = r"google_finance"   # A 文件夹路径
folder_D = r"NYSE"   # D 文件夹路径
file_B = r"NYSE.csv" # B.csv 路径
file_C = r"NYSE_missing.csv" # C.csv 路径

# 读取公司名单
companies_B = set(pd.read_csv(file_B, header=None)[0].astype(str).str.strip())
companies_C = set(pd.read_csv(file_C, header=None)[0].astype(str).str.strip())

# 取差集：在B里面但不在C里面
target_companies = companies_B - companies_C

# 遍历A文件夹
for file in os.listdir(folder_A):
    if file.endswith(".csv") and file.startswith("NYSE_"):
        parts = file.split("_")
        if len(parts) >= 3:  # 格式 NASDAQ_XXX_30Y.csv
            company = parts[1]
            if company in target_companies:
                src = os.path.join(folder_A, file)
                dst = os.path.join(folder_D, file)
                shutil.move(src, dst)
                # print(f"已移动: {file}")

print("完成！")
```
